#### 반복문으로 2차원 리스트의 요소를 모두 출력하기

* for 반복문을 한 번만 사용하기

```python
a = [[10, 20], [30, 40], [50,60]]

for x, y in a:
    print(x, y)
```

in 앞에 변수의 개수는 2차원 리스트에서 가로 크기(안쪽 리스트의 요소 개수)와 일치해야 함.

* for 반복문을 두 번 사용하기

```python
a = [[10, 20], [30, 40], [50, 60]]

for i in a:
    for j in i:
        print(j, end=' ')
    print()
```

for i in a:는 전체 리스트에서 가로 한 줄씩 꺼내 온다(안쪽 리스트를 통째로 꺼내 옴). <br/>다시 for j in i:와 같이 가로 한 줄(안쪽 리스트) i에서 요소를 하나씩 꺼내면 된다.

* for와 range 사용하기

```python
a = [[10, 20], [30, 40], [50, 60]]

for i in range(len(a)):            # 세로 크기
    for j in range(len(a[i])):     # 가로 크기
        print(a[i][j], end = ' ')
    print()
```

* while 반복문을 한 번 사용하기

```python
a = [[10, 20], [30, 40], [50, 60]]

i = 0
while i < len(a):    # 반복할 때 리스트의 크기 활용(세로 크기)
    x, y = a[i]      # 요소 두 개를 한꺼번에 가져오기
    print(x, y)
    i += 1           # 인덱스를 1 증가시킴
```

* while 반복문을 두 번 사용하기

```python
a = [[10, 20], [30, 40], [50, 60]]
 
i = 0
while i < len(a):           # 세로 크기
    j = 0
    while j < len(a[i]):    # 가로 크기
        print(a[i][j], end=' ')
        j += 1              # 가로 인덱스를 1 증가시킴
    print()
    i += 1                  # 세로 인덱스를 1 증가시킴
```

#### 반복문으로 리스트 만들기

* for 반복문으로 1차원 리스트 만들기

```python
a = []    # 빈 리스트 생성
 
for i in range(10):
    a.append(0)    # append로 요소 추가
 
print(a)
```

* for 반복문으로 2차원 리스트 만들기

```python
a = []    # 빈 리스트 생성
 
for i in range(3):
    line = []              # 안쪽 리스트로 사용할 빈 리스트 생성
    for j in range(2):
        line.append(0)     # 안쪽 리스트에 0 추가
    a.append(line)         # 전체 리스트에 안쪽 리스트를 추가
 
print(a)
```

* 리스트 표현식으로 2차원 리스트 만들기

```python
a = [[0 for j in range(2)] for i in range(3)]
a
```

> 먼저 [0 for j in range(2)]로 0을 2번 반복하여 [0, 0]으로 만들고 다시 for i in range(3)으로 [0, 0]을 3번 반복하여 [[0, 0], [0, 0], [0, 0]]으로 만든다.

```python
a = [[0] * 2 for i in range(3)]
a
```

>  0이 1개 들어있는 리스트 [0]에 2를 곱하면 [0, 0]이 되는데, 이걸 for i in range(3)으로 3번 반복해주면 [[0, 0], [0, 0], [0, 0]]이 된다.

* 톱니형 리스트 만들기

```python
a = [3, 1, 3, 2, 5]    # 가로 크기를 저장한 리스트
b = []    # 빈 리스트 생성
 
for i in a:      # 가로 크기를 저장한 리스트로 반복
    line = []    # 안쪽 리스트로 사용할 빈 리스트 생성
    for j in range(i):    # 리스트 a에 저장된 가로 크기만큼 반복
        line.append(0)
    b.append(line)        # 리스트 b에 안쪽 리스트를 추가
 
print(b)
```

리스트 a를 for로 반복하면 가로 크기를 꺼내면서 5번 반복<br/>
for 반복문 안에서 다시 for로 꺼낸 가로 크기 i만큼 반복하면서 append로 요소를 추가<br/>
바깥쪽 반복문에서 리스트 b에 안쪽 리스트 line을 추가

```python
a = [[0] * i for i in [3, 1, 3, 2, 5]]
a
```

* 2차원 리스트의 할당과 복사 알아보기

```python
a = [[10, 20], [30, 40]]
b = a
b[0][0] = 500
print(a)
print(b)
```

> 2차원 리스트를 만든 뒤 다른 변수에 할당하고, 요소를 변경해보면 두 리스트 모두 반영

```python
a = [[10, 20], [30, 40]]
b = a.copy()
b[0][0] = 500
print(a)
print(b)
```

> 리스트 a를 copy 메서드로 b에 복사한 뒤 b의 요소를 변경해보면 리스트 a와 b에 모두 반영

```python
a = [[10, 20], [30, 40]]
import copy             # copy 모듈을 가져옴
b = copy.deepcopy(a)    # copy.deepcopy 함수를 사용하여 깊은 복사
b[0][0] = 500
print(a)
print(b)
```

> 2차원 이상의 다차원 리스트는 리스트를 완전히 복사하려면 copy 메소드 대신 copy 모듈의 deepcopy 함수를 사용해야 함.

* 연습문제: 3차원 리스트 만들기
높이 2, 세로 크기 4, 가로 크기 3인 3차원 리스트를 만드시오.(리스트 표현식 사용)

```python
a = [[[0 for col in range(3)] for row in range(4)] for depth in range(2)]
a
```

> 3차원 리스트는 높이 x 가로 x 세로 형태로 이루어져 있음.

* 리스트 = [[[값, 값], [값, 값]], [[값, 값], [값, 값]], [[값, 값], [값, 값]]]
* 리스트[높이인덱스][세로인덱스][가로인덱스]
* 리스트[높이인덱스][세로인덱스][가로인덱스] = 값

* 연습문제: 지뢰찾기

```python
# m*n input 
row, col = map(int, input("> 입력: ").split())

# matrix 리스트 생성 및 input 값 세로 범위로 리스트 추가 
matrix = []
for i in range (row):
     matrix.append(list(input()))

# 세로 가로 이중 포문 및 조건문으로 *(지뢰)과 같으면 그대로 출력        
    
for i in range(row):
    for j in range(col):
        if matrix[i][j] == '*':
            print(matrix[i][j], sep = '', end = '')
# 그렇지 않으면 
# cnt = 0 초기 값 설정
        else:
            cnt = 0
# 세로 가로 -1  세로 가로 +2 로 범위를 설정으로 기준점에 인접한 8개 인덱스 범위 지정            
            for x in range(i - 1, i + 2):
                for y in range(j - 1, j + 2):
# if not() 으로 예외상황 필터
# [0][0] 일 때, i,j -1 인 경우 음수 발생
# m*n = 5*5, [5][5] 일 때, i,j +2 인 경우 인덱스 범위를 초과하는 범위 발생
# if not()이 아닌 경우 if 중첩 조건문으로 x y 값이 *과 같으면 cnt에 +1씩 추가
# 해당 인덱스 cnt 합계 출력
                    if not (x < 0 or y < 0 or x >= row or y >= col):
                        if matrix[x][y] == "*":
                            cnt += 1           
            print(cnt, sep = '', end = '')
# 모든 반복이 끝나면 한번에 출력                
    print()

                        
                       
                            
                
                        
                       
     
                    
                    

    
```
