# 17.1 while 반복문 사용하기


while 반복문은 조건식으로만 동작하며 반복할 코드 안에 조건식에 영향을 주는 변화식이 들어갑니다.

```python
i = 0
while i < 10:
    print('ddd',i)
    i += 1
```

* += 덧셈 후 대입
* -= 뺄셈 후 대입
* *= 곱셈 후 대입
* /= 나눗셈 후 대입
* %= 나머지 구한 후 대입
* **= 제곱 후 대입

초기식 <br/>
<t/>while 조건식:<br/>
     반복할 코드<br/>
     변화식

### 17.1.1  초깃값을 1부터 시작하기


```python
i = 1
while i <= 10:
    print('dd', i)
    i += 1
```

### 17.1.2  초깃값을 감소시키기


```python
i = 10
while i > 0:
    print('dd', i)
    i -= 1
```

### 17.1.3  입력한 횟수대로 반복하기


```python
count = int(input("반복할 횟수를 입력하세요: "))

i = 0
while i < count:
    print('dd',i)
    i += 1
```

```python
count = int(input('반복 횟수 입력: '))

while count > 0:
    print('ddd',count)
    count -= 1
    
```

### 17.5 연습문제: 변수 두 개를 다르게 반복하기


다음 소스 코드를 완성하여 정수 2 5, 4 4, 8 3, 16 2, 32 1이 각 줄에 출력되게 만드세요. <br/>while에 조건식은 두 개 지정하고, 두 변수를 모두 변화시켜야 합니다.

```python
i = 2
j = 5

while i <= 32 or j >= 1:
    print(i, j)
    i *= 2
    j -= 1
```

### 17.6 심사문제: 교통카드 잔액 출력하기


표준 입력으로 금액(정수)이 입력됩니다.<br/> 1회당 요금은 1,350원이고, 교통카드를 사용했을 때마다의 잔액을 각 줄에 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다). <br/> 단, 최초 금액은 출력하지 않아야 합니다. 그리고 잔액은 음수가 될 수 없으며 잔액이 부족하면 출력을 끝냅니다.

```python
count = int(input('양의 정수 입력: '))

while count >= 1350:
    count = count - 1350
    print(count)
```
