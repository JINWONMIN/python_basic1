 * 리스트에 요소 추가하기

```python
a = [10, 20, 30]
a.append(500)
a
```

```python
a = []
a.append(10)
a
```

* 리스트 안에 리스트 추가하기

```python
a = [10, 20, 30]
a.append([500, 600])
print(a)
len(a)
```

* 리스트의 특정 인덱스에 요소 추가하기

insert(인덱스, 요소)

```python
a = [10, 20, 30]
a.insert(2, 500)
print(a)
len(a)
```

> insert에서 자주 사용되는 패턴 2가지
* insert(0, 요소): 리스트의 맨 처음에 요소를 추가
* insert(len(리스트), 요소): 리스트 끝에 요소를 추가

```python
a = [10, 20, 30]
a.insert(0, 500)
print(a)

b = [10, 20, 30]
b.insert(len(b), 500)  # b.append(500)과 동일
print(b)
```

* insert(인덱스, [리스트])

```python
a = [10, 20, 30]
a.insert(1, [500, 600])
print(a)
```

* 슬라이싱

```python
a = [10, 20, 30]
a[1:1] = [500, 600]
a
```

* 리스트에서 요소 삭제하기
<br/>pop: 마지막 요소 또는 특정 인덱스의 요소를 삭제
<br/>remove: 특정 값을 찾아서 삭제

* 리스트에서 특정 인덱스의 요소를 삭제하기

```python
a = [10, 20, 30]
a.pop()
print(a)
```

```python
a = [10, 20, 30]
a.pop(1)
a
```

* 리스트에서 특정 값을 찾아서 삭제하기

```python
a = [10, 20, 30]
a.remove(20)
a
```

```python
a = [10, 20, 30, 20]
a.remove(20)
a
```
